```python
def string2array(direction):
    lenght = len(direction.replace("-",''))
    arr = np.empty(lenght)
    j = False
    i = 0
    k = 0 
    while i < len(direction): 
        char = direction[i]
        if char == "-":
            j = True
            i += 1
            k += 1
            continue
        if j:
            arr[i-k] = -int(char)
            j = False
        else:
            arr[i-k] = int(char)
        i += 1 
    return arr
# %timeit string2array("001")
```

```python

def findPoints(listofps, allpoints, Anti):
    for ps in listofps:
        indx = findindex(ps,allpoints)    
        if (indx + 1 and porownajPunkty(allpoints[:,indx],ps)) == Anti:
            return False
    return True 

def findAntiSym_InnerLoop(Matrix, allpoints, vacancies):
    for vac in vacancies:
        if not findPoints(listadous(Matrix,vac), allpoints.T, Anti=True):
            return False
    return True

def findAntiSym(matrixes, allpoints, vacancies):
    allsymmerties = makelist("h")
    if vacancies.shape == (4,):
        vacancies = np.array([vacancies])
    possymmerties = []    
    for el0, el1 in allsymmerties:
        if findAntiSym_InnerLoop(matrixes[el0][el1], allpoints, vacancies):
            possymmerties.append((el0, el1))           
    return possymmerties 

def findSym_innerLoop(Matrix, allpoints, vacancies = []):
    if not vacancies:
        vacancies = allpoints
    for p in allpoints:   
        if not findPoints(listadous(Matrix,p), allpoints.T, Anti = False):
            return False
    return True

def findSym(matrixes, allpoints, vacancies):
    symmerties = []
    possym = findAntiSym(matrixes, allpoints, vacancies)
    for el0, el1 in possym:                
        if findSym_innerLoop(matrixes[el0][el1], allpoints):
            symmerties.append((el0, el1))
    return symmerties
```
