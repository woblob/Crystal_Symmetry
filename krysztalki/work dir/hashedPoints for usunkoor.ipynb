```python
class HashedPoints():
    """
    SET = set(((3,4),(2,4),(2,2),(2,3)))
    old = set(((1,1),(1,2),(1,3)))
    new = set(((1,1),(2,2),(2,3)))
    dodanie_puntkow = old - new
    odjecie_puntkow = new - old & new

    SET = (SET | dodanie_puntkow) - odjecie_puntkow

    print(f'{SET=}\n{dodanie_puntkow=}\n{odjecie_puntkow=}')# vakancje 

    """
    def __init__(self, array):
        self.SET = HashedPoints.to_set(array)
        self.present_vacancies = set()
    
    @staticmethod
    def to_set(array):
        return {tuple(point) for point in array}
    
    
    def update(self, imput):
        new_vacancies = HashedPoints.to_set(imput)
        dodanie_puntkow = self.present_vacancies - new_vacancies
        odjecie_puntkow = new_vacancies - (self.present_vacancies & new_vacancies)
        
        self.SET = (self.SET | dodanie_puntkow) - odjecie_puntkow
        self.present_vacancies = new_vacancies
    
    def find_symmetries_for_points(self, transformed_vacancies):
        symmetry_mask = np.full(len(transformed_vacancies), True)
        for tr, points in enumerate(transformed_vacancies):
            raise 'check if points.shape[0] > 1'
            for point in points:
                if tuple(point) in self.SET:
                    symmetry_mask[tr] = False
                    break

        return symmetry_mask

    def find_symmetries_for_points(self, transformed_vacancies):
        symmetry_mask = np.empty(len(transformed_vacancies), dtype=bool)
        for tr, points in enumerate(transformed_vacancies):
#             raise 'check if points.shape[0] > 1'
#             symmetry_mask[tr] = any(True for point in points if tuple(point) in self.SET)
            symmetry_mask[tr] = tuple(points) in self.SET
        return symmetry_mask
                    
    
    
    def __str__(self):
        return str(self.SET)
    
    def __repr__(self):
        return repr(self.SET)
```
