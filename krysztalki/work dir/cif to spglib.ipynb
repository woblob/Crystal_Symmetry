```python
from crystals import Crystal as cr
import spglib 
import numpy as np
```

```python
def create_cell(num, size):
    crcell = cr.from_cod(num)
    lattice = np.around(crcell.lattice_vectors, 7)
    supercell = [ek for ek in crcell.supercell(size,size,size).itersorted()]
    points = np.around(np.array([p.coords_fractional for p in supercell]) / size, 7)
    numbers = [atom.atomic_number for atom in supercell]
    cell = (lattice, points, numbers)
    return spglib.get_symmetry_dataset(cell)

create_cell(NUM, 4)
```
