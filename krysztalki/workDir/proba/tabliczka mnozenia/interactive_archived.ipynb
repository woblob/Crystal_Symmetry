```python
def _find_base(matrix, all_bases):
    for i, base in enumerate(all_bases):
        cond = np.array_equal(base, matrix)
        if cond:
            return i
    notfound = -1 
    return notfound

def find_matrix(matrix, bases):
    mat = matrix[:-1, :-1]
    return _find_base(mat, bases)

def find_translation(matrix, bases):
    trans = matrix[:-1, -1]
    return _find_base(trans, bases)

def separate_matrix(matrix):
    rot = matrix[:-1, :-1]
    trans = matrix[:-1, -1]
    base = np.zeros((4, 4))
    
    rotation = base.copy()
    rotation[:-1, :-1] = rot
    rotation[-1, -1] = 1    
    
    translation = base.copy()
    translation[:-1, -1] = trans
    
    return rotation, translation
```
